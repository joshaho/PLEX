```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
FullBudget=pd.read_excel('PLEX2018Budget.xlsx', sep=',')
FullBudget
```

```python
#Calculate Total Income/Expenses

#Income
incomes=pd.DataFrame(FullBudget['Line Total'][FullBudget['Expense/Income']=='Income'])
plusses=incomes.sum().values[0]
#Expenses
expenses=pd.DataFrame(FullBudget['Line Total'][FullBudget['Expense/Income']!='Income'])
minuses=expenses.replace('[\$,]', '', regex=True).astype(float).sum().values[0]
[plusses,minuses]
```

```python
#By Category

DataDump=pd.ExcelWriter('PLEX 2018 Breakdowns by Category.xlsx')

#Calculate Net Budgets Per Category
Categories=FullBudget['Category'].unique() #Figure out how to generate this automatically
Net=FullBudget['Line Total'].replace('[\$,]', '', regex=True).astype(float).sum()
TotalDataFrame=pd.DataFrame({'Category': ['Event Net'], 'Total': ['$'+str(Net)]})
for category in Categories:
    categorybudget=FullBudget[FullBudget['Category']==category]
    #Calculate Net Total for Category
    total=categorybudget['Line Total'].replace('[\$,]', '', regex=True).astype(float).sum()
    total_row= ['','','','','','','','','Category Total:','$'+str(total),'']
    categorybudget.loc[len(categorybudget)] = total_row
    categorybudget.to_excel(DataDump, sheet_name=category, index=False)
    TotalDataFrame.loc[len(TotalDataFrame)]=[category, '$'+str(total)]
TotalDataFrame.to_excel(DataDump, sheet_name='Category Summary', index=False)
FullBudget.to_excel(DataDump, sheet_name='Full Budget', index=False)
DataDump.save()
    
```

```python
# By Team
DataDump=pd.ExcelWriter('PLEX 2018 Breakdowns by Team.xlsx')
BreakdownColumn='Team'
#Calculate Net Budgets Per Category
Categories=FullBudget[BreakdownColumn].unique() #Figure out how to generate this automatically
Net=FullBudget['Line Total'].replace('[\$,]', '', regex=True).astype(float).sum()
TotalDataFrame=pd.DataFrame({BreakdownColumn: ['Event Net'], 'Total': ['$'+str(Net)]})
for category in Categories:
    categorybudget=FullBudget[FullBudget[BreakdownColumn]==category]
    #Calculate Net Total for Category
    total=categorybudget['Line Total'].replace('[\$,]', '', regex=True).astype(float).sum()
    total_row= ['','','','','','','','',category+' Total:','$'+str(total),'']
    categorybudget.loc[len(categorybudget)] = total_row
    categorybudget.to_excel(DataDump, sheet_name=category, index=False)
    TotalDataFrame.loc[len(TotalDataFrame)]=[category, '$'+str(total)]
TotalDataFrame.to_excel(DataDump, sheet_name=BreakdownColumn+' Summary', index=False)
FullBudget.to_excel(DataDump, sheet_name='Full Budget', index=False)
DataDump.save()
```
